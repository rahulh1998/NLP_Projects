```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential,save_model
from keras.layers import Dense,Flatten,Conv1D,GlobalMaxPool1D,Dropout,Embedding,LSTM
import itertools
import os
from tensorflow.keras.utils import to_categorical
```

```python
for file in os.listdir('data/'):
    with open('data/'+file ,encoding='utf-8') as f:
        text_data =[line.strip() for line in  f.readlines()]
```

```python
text_data
```

```python
# Tokenizing the text data

token = Tokenizer()
token.fit_on_texts(text_data)
```

```python
token.word_counts
```

```python
vocab_size = len(token.word_counts) + 1
```

```python
token.word_index
```

```python
text = token.texts_to_sequences(text_data)
text
```

# Preparing training data

```python
data = []
for d in text:
    if len(d) > 1:
        for i in range(2,len(d)):
            data.append(d[:i])
            print(d[:i])
```

# Applying Padding

```python
max_length = 50
sequences = pad_sequences(data,maxlen= max_length,padding='pre')
sequences
```

# Creating X & y

```python
X = sequences[ : , :-1]
y = sequences[ : , -1 ]
```

```python
y= to_categorical(y,num_classes=vocab_size)
```

```python
seq_len = X.shape[1]
```

# LSTM Model

```python
model = Sequential()
model.add(Embedding(vocab_size,50,input_length=seq_len))
model.add(LSTM(100, return_sequences=True))
model.add(LSTM(100))
model.add(Dense(100,activation='relu'))
model.add(Dense(vocab_size,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',metrics=['accuracy'],loss = 'categorical_crossentropy')
```

```python
model.fit(X,y,epochs=100,batch_size=32)
```

# Poetry Generation

```python
poetry_length = 15
def generate_poetry(seed_text, n_lines):

    for i in range(n_lines):
        text = []
        for _ in range(poetry_length):
            encoded = token.texts_to_sequences([seed_text])
            encoded = pad_sequences(encoded, maxlen = 49, padding='pre')

            y_pred = np.argmax(model.predict(encoded), axis=-1)

            predicted_word = ""
            for word, index in token.word_index.items():
                if index == y_pred:
                    predicted_word = word
                    break

            seed_text = seed_text + ' ' + predicted_word
            text.append(predicted_word)

        seed_text = text[-1]
        text = ' '.join(text)
        print(text)
```

```python
seed_text = "Must be love on the brain"
generate_poetry(seed_text, 10)
```

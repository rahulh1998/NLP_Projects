The model used is called facebook/bart-large-cnn and has been developed by Facebook. It contains 1024 hidden layers and 406M parameters and has been fine-tuned using CNN, a news summarization dataset.

```python
# Importing libraries
from transformers import pipeline
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
import requests
import re
```

```python
# for getting the data from wikipedia
response = requests.get("https://en.wikipedia.org/wiki/Lionel_Messi")

# for displaying all the data with tags
response.text[:500]
```

```python
# Creating BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# Extract all the data in <p> tags 
p_tags = soup.find_all('p')
```

```python
# Remove any tags between p tags
for p in p_tags:
    next_tag = p.find_next_sibling()
    while next_tag and next_tag.name != 'p':
        next_tag.unwrap()
        next_tag = p.find_next_sibling()
```

```python
# Extract the text content of p tags
for p in p_tags:
    print(p.text)
```

```python
type(p_tags)
```

```python
text_data = []
for p in p_tags:
    text_data.append(p.text)
    
text_data = " ".join(text_data)
```

```python
text_data
```

```python
# Cleaning the data 
text_data = re.sub(r'\[\d+\]'," ",text_data)
text_data = re.sub(r'\n','',text_data)
text_data = re.sub(r'\xa0','',text_data)
text_data
```

```python
# Total words
len(text_data.split())
```

# Summarization

```python
from transformers import pipeline

summ = pipeline('summarization',model="facebook/bart-large-cnn")
res = summ(text_data[:4500], max_length = 500)[0]['summary_text']
res
```
